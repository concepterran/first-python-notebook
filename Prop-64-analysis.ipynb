A Prop 64 Analysis

```python
import pandas as pd
```

Reading in the list of propositions and committees from CSV

```python
props = pd.read_csv("https://first-python-notebook.readthedocs.io/_static/committees.csv")
```

Reading in the list of contributions to ballot measures

```python
contribs = pd.read_csv("https://first-python-notebook.readthedocs.io/_static/contributions.csv")
```

Making the list to be prop 64 isolated

```python
prop64 = props[props.prop_name == "PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE."]
```

```python
prop64.head()
```

```python
merged = pd.merge(prop64, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

```python
support = merged[merged.committee_position =='SUPPORT']
```

```python
oppose = merged[merged.committee_position =='OPPOSE']
```

```python
oppose.amount.sum()
```

```python
support.amount.sum()
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
